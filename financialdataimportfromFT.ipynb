```python
import requests 
import lxml.html
import json
import re
import ipywidgets as widgets
from datetime import datetime
from IPython.display import display

def getHistory( ftid, fromdate, todate  ):
        
        x = requests.get("https://markets.ft.com/data/equities/ajax/get-historical-prices?startDate=" 
            + str(fromdate.year) + "%2F" 
            + str(fromdate.month) + "%2F" 
            + str(fromdate.day) + "&endDate=" 
            + str(todate.year) + "%2F" 
            + str(todate.month) + "%2F" 
            + str(todate.day) + "&symbol=" 
            + str(ftid))
        responsetext = lxml.html.fromstring(json.loads(x.content.decode())["html"]).text_content()
        return re.findall("(\w+, \w+ \d\d, \d\d\d\d)(\w+, \w+ \d\d, \d\d\d\d)([\d,]+\.\d\d)([\d,]+\.\d\d)([\d,]+\.\d\d)([\d,]+\.\d\d)",responsetext)
   
```

```python
display(widgets.DatePicker(
    description='From Date',
    disabled=False
))
```

```python
fromdate = datetime.strptime("01/01/2018","%d/%m/%Y")
todate = datetime.strptime("31/12/2018","%d/%m/%Y")
ftid = 35662827 #(LifeStrategy 100% Equity Fund - Accumulation)

getHistory( 35662827, fromdate ,todate )
```
